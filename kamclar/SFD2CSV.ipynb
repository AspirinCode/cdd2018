```python
from rdkit import Chem
from rdkit.Chem import rdinchi
from rdkit.Chem import Descriptors
from rdkit.Chem import Lipinski
from rdkit.Chem import rdMolDescriptors as rdescriptors
import csv
```

```python
drugbank_input = Chem.SDMolSupplier('../data/drugbank.sdf')
drugbank = [m for m in drugbank_input if m]
```

```python
with open ('drugbank.csv', 'w') as csvFile:
    drugbankWriter = csv.writer(csvFile, delimiter = ';', quotechar = '"', quoting = csv.QUOTE_MINIMAL)
    drugbankWriter.writerow(['inchikey', 'smiles', 'molwt', 'logP', 'HBA', 'HBD'])
    for drug in drugbank:
        try:
            inchi = rdinchi.MolToInchi(drug)[0]
            inchikey = rdinchi.InchiToInchiKey(inchi)
        except ValueError:
            print(Chem.MolToSmiles(drug)[0])
            inchi = None
            inchikey = None
        drugbankWriter.writerow([inchikey, Chem.MolToSmiles(drug),
                                Descriptors.MolWt(drug),
                                Descriptors.MolLogP(drug),
                                rdescriptors.CalcNumLipinskiHBA(drug),
                                rdescriptors.CalcNumLipinskiHBD(drug),])
```

```python
with open('drugbank.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=';',
                                 quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for x in reader:
        print(x)
```

```python
m=0
pocetKruhu = 0
for drug in drugbank:
    if rdescriptors.CalcNumAromaticRings(drug) != 0:
        m+=1
        pocetKruhu += rdescriptors.CalcNumAromaticRings(drug)
    prumKruhy = pocetKruhu / m        
    
print(m, pocetKruhu, int(prumKruhy))
```

```python
aromaticDrugs = [m for m in drugbank if Lipinski.NumAromaticRings(m) > 0]
print(len(aromaticDrugs))
```

```python
pattern = Chem.MolFromSmarts('c1ccccc1')
```

```python
notMatchingdrugs = [m for m in aromaticDrugs if not m.HasSubstructMatch(pattern)]
```

```python
kBenzoova = Chem.MolFromSmarts('c1ccccc1C(=O)O')
```

```python
obsahujekB = [m for m in aromaticDrugs if m.HasSubstructMatch(kBenzoova)]
```

```python
obsahujeVicKB = [m for m in aromaticDrugs if len(m.GetSubstructMatches(kBenzoova)) > 1]
```

```python
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import SVG
```

```python
drawer = rdMolDraw2D.MolDraw2DSVG(333, 333)
mol = obsahujekB[1]
drawer.DrawMolecule(mol,
                    highlightAtoms=mol.GetSubstructMatches(kBenzoova)[0])
drawer.FinishDrawing()
SVG(drawer.GetDrawingText().replace('svg:',''))
```

```python
drawer = rdMolDraw2D.MolDraw2DSVG(333, 333)
mol = obsahujeVicKB[1]
atoms2highlight = set()
for match in mol.GetSubstructMatches(kBenzoova):
    for atom in match:
        atoms2highlight.add(atom)
drawer.DrawMolecule(mol,
                    highlightAtoms=atoms2highlight)
drawer.FinishDrawing()
SVG(drawer.GetDrawingText().replace('svg:',''))
```
